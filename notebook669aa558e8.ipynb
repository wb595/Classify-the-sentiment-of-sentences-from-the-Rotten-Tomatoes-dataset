```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
```

```python
train_data = pd.read_csv('/kaggle/input/sentiment-analysis-on-movie-reviews/train.tsv.zip', sep='\t')
print(train_data.head())   

test_data = pd.read_csv('/kaggle/input/sentiment-analysis-on-movie-reviews/test.tsv.zip', sep='\t')
print(test_data.head())   

```

```python
test_data = pd.read_csv('test.tsv', sep='\t')
print(test_data.head())
```

```python
vectorizer = CountVectorizer(ngram_range=(1,3),  
                             max_features = 150000) 

corpus_train = train_data['Phrase']  
corpus_test = test_data['Phrase']  

vectorizer.fit(pd.concat([corpus_train,corpus_test]))
print(vectorizer.get_feature_names()[:10])

X_train = vectorizer.transform(corpus_train)    
X_test = vectorizer.transform(corpus_test)   

y_train = list(train_data['Sentiment'])

print(type(X_train))
print(X_train[:1])
print(y_train[:5])
```

```python
model = LogisticRegression(max_iter=1000000)
model.fit(X_train, y_train)

y_test = model.predict(X_test)
#y_test = [4]*66292
```

```python
output = pd.DataFrame({'PhraseId': test_data.PhraseId,
                       'Sentiment': y_test})
output.to_csv('my_submission.csv', index=False)
```
